```python
import requests
import pandas as pd

def fetch_crypto_data(crypto_id='bitcoin', vs_currency='usd', days='max'):
    url = f'https://api.coingecko.com/api/v3/coins/{crypto_id}/market_chart'
    params = {'vs_currency': vs_currency, 'days': days}
    response = requests.get(url, params=params)

    # Check the response status
    if response.status_code == 200:
        data = response.json()
        # Print the response JSON to see its structure
        print("API response:", data)
        
        # Check for the 'prices' key
        if 'prices' in data:
            prices = data['prices']
            df = pd.DataFrame(prices, columns=['timestamp', 'price'])
            df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
            return df
        else:
            print("Error: 'prices' key not found in the API response.")
            return None
    else:
        print(f"Error: API request failed with status code {response.status_code}")
        return None

# Fetch data
bitcoin_data = fetch_crypto_data()
if bitcoin_data is not None:
    print(bitcoin_data.head())
else:
    print("Data could not be retrieved.")
```

```python
import pandas as pd
import os

# Define the folder path
folder_path = "C:/Users/USER/Downloads/cryptocurrency"

# Define file paths for each cryptocurrency
files = {
    "Bitcoin": os.path.join(folder_path, "bitcoin_historical_data.csv"),
    "Ethereum": os.path.join(folder_path, "ethereum_historical_data.csv"),
    "Binance_Coin": os.path.join(folder_path, "Binance_Coin_historical_data.csv"),
    "Dogecoin": os.path.join(folder_path, "Dogecoin_historical_data.csv"),
    "USDT": os.path.join(folder_path, "USDT_historical_data.csv")
}

# Initialize an empty list to store each dataframe
dataframes = []

# Loop through each file, add a 'Coin' column, and append to the list
for coin_name, file_path in files.items():
    # Load the data
    df = pd.read_csv(file_path)
    
    # Add the 'Coin' column
    df['Coin'] = coin_name
    
    # Append to the list of dataframes
    dataframes.append(df)

# Concatenate all dataframes into one
combined_df = pd.concat(dataframes, ignore_index=True)

# Save the combined dataframe
combined_file_path = os.path.join(folder_path, "all_cryptocurrencies_data.csv")
combined_df.to_csv(combined_file_path, index=False)

print(f"Combined file saved at: {combined_file_path}")
```
